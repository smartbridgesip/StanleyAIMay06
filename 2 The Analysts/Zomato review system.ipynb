```python
import vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
zomato['Text Review Score'] = zomato['Rating text'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
```

```python
zomato
```

```python
positive = len(zomato[zomato['Text Review Score'] >=0.05])
neutral = len(zomato[(zomato['Text Review Score'] >-0.05) & (zomato['Text Review Score']<0.05)])
negative = len(zomato[zomato['Text Review Score']<=-0.05])
```

```python
positive,neutral, negative
```

```python
zomato.loc[zomato['Text Review Score'] >=0.05,"Text Sentiment Label"] ="positive"
zomato.loc[(zomato['Text Review Score'] >-0.05) & (zomato['Text Review Score']<0.05),"Text Sentiment Label"]= "neutral"
zomato.loc[zomato['Text Review Score']<=-0.05,"Text Sentiment Label"] = "negative"
```
